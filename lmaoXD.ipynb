# Model Creation

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder 
import pickle
```

```python
le = preprocessing.LabelEncoder()
```

## Label encoding the categorical columns: State and county

```python
le.fit(df['State'])
le_State_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(le_State_mapping)
df['State'] = le.transform(df['State'])
```

```python
le.fit(df['County'])
le_County_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(le_County_mapping)
df['County'] = le.transform(df['County'])
```

```python
mapping_dict = {
    'State': le_State_mapping,
    'County': le_County_mapping
}
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
file_path = 'encoded_values.pkl'

with open(file_path, 'wb') as file:
    pickle.dump(mapping_dict, file)
```

## Scaling and splitting the dataset

```python
scaler = StandardScaler()
X = df.drop("90th Percentile AQI", axis = 1)
y = df["90th Percentile AQI"]
scaler.fit(X)
X = scaler.transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)
```

```python
import xgboost as xgb
dtrain_reg = xgb.DMatrix(X_train, y_train)
dtest_reg = xgb.DMatrix(X_test, y_test)
params = {"objective": "reg:squarederror", "tree_method": "gpu_hist"}
xgbModel = xgb.train(
   params=params,
   dtrain=dtrain_reg
)
```

```python
y_pred = xgbModel.predict(dtest_reg)

xgbMAE = mean_absolute_error(y_test, y_pred)

print("\nMAE = ", xgbMAE)
```

# ANN Model #1

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
# Define the model
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[17, ]),  # input_dim is the number of features in your dataset
    layers.Dense(64, activation='relu'),
    layers.Dense(1)  # Output layer with a single node for regression
])
```

```python
model.compile(optimizer='adam', loss='mean_absolute_error')
```

```python
model.fit(X_train, y_train, epochs=50)
```

```python
y_pred = model.predict(X_test)
```

```python
print(model.evaluate(X_test, y_test))
```

```python
mae = mean_absolute_error(y_test, y_pred)
print(mae)
```
